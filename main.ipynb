````python
import pandas as pd
````

## Открываем файл

````python
file = pd.read_csv("pg_stat_statements_202412061313.csv", sep=",")
file.head()
````

## Сортируем запросы по общему времени выполнения.

````python
sorting_by_total_time = file.sort_values(by="total_time", axis=0, ascending=False)
sorting_by_total_time.head()
````

## Отбираем два самых долгих запроса:

````python
num_query = 2
time_longest = sorting_by_total_time["total_time"][:num_query].sum()
time_all = sorting_by_total_time["total_time"].sum()
res = (time_longest / time_all) * 100
print(f"Процент занимаемого времени: {res:.2f}%")
````

## Сортируем запросы по среднему времени выполнения.

````python
sorting_by_mean_time = file.sort_values(by="mean_time", axis=0, ascending=False)
sorting_by_mean_time.head()
````
